```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
can = load_breast_cancer()
```

```python
can
```

```python
df = pd.DataFrame(np.c_[can.data, can.target], columns = list(can.feature_names)+['target'])
```

```python
df
```

```python
x = df.iloc[ : , :-1]
y = df.iloc[ : , -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(x_train,y_train)
```

```python
nb.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()

mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)
```

```python
df = pd.read_csv('C:/Users/DELL/Desktop/DATA/Python/Weekend/Machine Learning/09 April/spam.csv', encoding = 'latin-1')
```

```python
df
```

```python
df.drop(columns = ['Unnamed: 2','Unnamed: 3','Unnamed: 4'], inplace = True)
```

```python
df
```

```python
df['v1'] = df['v1'].apply(lambda x : 1 if x == 'spam' else 0)
```

```python
df
```

```python
x = df['v2']
y = df['v1']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
x = cv.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
nb.fit(x_train,y_train)
```

```python
nb.score(x_test,y_test)
```

```python
email = ([
    'you have won lottery with offer',
 '20% discount'])
```

```python
email = cv.transform(email)
```

```python
nb.predict(email)
```
