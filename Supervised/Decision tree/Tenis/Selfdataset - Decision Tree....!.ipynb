```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Weather'] = le.fit_transform(df.Weather)
df['Parents'] = le.fit_transform(df.Parents)
df['Money'] = le.fit_transform(df.Money)
df['Decision'] = le.fit_transform(df.Decision)
```

```python
df
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(25,25))

tree.plot_tree(model,feature_names=X.columns,filled=True,class_names=df1.Decision.unique())
plt.show()
```
