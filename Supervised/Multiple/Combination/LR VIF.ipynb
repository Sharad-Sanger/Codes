```python
def remove_outliers(df, a):
    
    q1 = np.percentile(df[a], 25)
    q3 = np.percentile(df[a], 75)
    iqr = q3 - q1

    lower_bound = q1 - (1.5 * IQR)
    upper_bound = q3 + (1.5 * IQR)
    
    df = (df[a] >= lower_bound) & (df[a] <= upper_bound)
    return df
```
