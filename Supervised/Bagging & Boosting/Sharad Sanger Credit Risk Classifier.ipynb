```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('C:\\Users\\DELL\\Desktop\\New folder (2)\\customer_data.csv')
df2 = pd.read_csv('C:\\Users\\DELL\\Desktop\\New folder (2)\\payment_data.csv')
```

```python
df1.info()
```

```python
df2.info()
```

```python
df = (df1.merge(df2, how = 'inner'))
df
```

```python
df.drop('id', axis = 1, inplace = True)
df.drop('update_date', axis = 1, inplace = True)
df.drop('report_date', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
col = df.columns
```

```python
for i in col:
    if df[i].dtypes=='object':
        df[i] = df[i].fillna(df[i].mode()[0])
    else:
        df[i] = df[i].fillna(df[i].mean())
```

```python
df.info()
```

```python
y = df.loc[ : , 'label']
```

```python
x = df.iloc[ : , 1 :]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, train_size=0.2)
```

# Bagging Techniques

```python
# RANDOM FOREST
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier(n_estimators=100,random_state=42)

rfc.fit(x_train,y_train)

y_pred=rfc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Bagged Decision Trees
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()

from sklearn.ensemble import BaggingClassifier

bc=BaggingClassifier(base_estimator=tree)

bc.fit(x_train,y_train)

y_pred=bc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# Extra Trees
```

```python
from sklearn.ensemble import ExtraTreesClassifier

et=ExtraTreesClassifier()

et.fit(x_train,y_train)

y_pred=et.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
