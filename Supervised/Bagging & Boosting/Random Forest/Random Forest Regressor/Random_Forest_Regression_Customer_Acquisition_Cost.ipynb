```python
#label Encoding Categorical non-numeric variables
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
media['media_type'] = le.fit_transform(media['media_type'])
media['store_state'] = le.fit_transform(media['store_state'])
media['store_city'] = le.fit_transform(media['store_city'])
media['store_type'] = le.fit_transform(media['store_type'])
media['brand_name'] = le.fit_transform(media['brand_name'])
media['avg. yearly_income'] = le.fit_transform(media['avg. yearly_income'])
media['houseowner'] = le.fit_transform(media['houseowner'])
media['occupation'] = le.fit_transform(media['occupation'])
media['member_card'] = le.fit_transform(media['member_card'])
media['education'] = le.fit_transform(media['education'])
media['gender'] = le.fit_transform(media['gender'])
media['marital_status'] = le.fit_transform(media['marital_status'])
media['sales_country'] = le.fit_transform(media['sales_country'])
media['promotion_name'] = le.fit_transform(media['promotion_name'])
media['food_family'] = le.fit_transform(media['food_family'])
media['food_department'] = le.fit_transform(media['food_department'])
media['food_category'] = le.fit_transform(media['food_category'])
```

```python
#visual Representation of the data, EDA
```

```python
X = media.drop(['cost'], axis=1)
y = media['cost']
```

```python
#Recursive Feature Selection
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split, StratifiedKFold

rfe = RFE(RandomForestRegressor(), n_features_to_select=10)
```

```python
rfe.fit(X, y)
```

```python
rfe.ranking_
```

```python
rfe.get_support()
```

```python
feature_selected = X.columns[rfe.get_support()]
feature_selected
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train = rfe.transform(X_train)
X_test = rfe.transform(X_test)
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
#evaluation
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#mean squared error
mse = mean_squared_error(y_test, pred)
print(mse)
```

```python
#mean absolute error
mae = mean_absolute_error(y_test, pred)
print(mae)
```

```python
#r2 score 
r2 = r2_score(y_test, pred)
print(r2)
```
