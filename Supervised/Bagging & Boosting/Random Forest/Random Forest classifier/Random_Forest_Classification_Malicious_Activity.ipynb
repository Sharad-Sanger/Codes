```python
#feature selection using RFECV 
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFECV 
from sklearn.model_selection import train_test_split, StratifiedKFold
```

```python
X = data.drop(['MALICIOUS_OFFENSE'], axis=1)
y = data['MALICIOUS_OFFENSE']
```

```python
rfe = RFECV(RandomForestClassifier(), cv=StratifiedKFold(3))
```

```python
rfe.fit(X, y)
```

```python
print(rfe.n_features_)
print(rfe.ranking_)
print(rfe.get_support())
```

```python
feature_selected = X.columns[rfe.get_support()]
feature_selected
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
X_train = rfe.transform(X_train)
X_test = rfe.transform(X_test)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

cf = classification_report(y_test, pred)
print(cf)
```

```python
c_f = confusion_matrix(y_test, pred)
print(c_f)
```

```python
import seaborn as sns

sns.heatmap(c_f)
```
