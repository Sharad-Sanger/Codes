```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

```python
y = df.pop("iris")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x = sc.fit_transform(df)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=11)
```

```python
pca.fit(x)
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(random_state=11,n_components=2)
```

```python
a = pca.fit_transform(x)
```

```python
a
```

```python
df2 = pd.DataFrame(a,columns=['Pc1','Pc2'])
```

```python
df2.head()
```

```python
df2['label'] =y
```

```python
df2.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='Pc1',y='Pc2',hue='label',data=df2)
```
