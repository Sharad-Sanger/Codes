```python
import tensorflow
```

```python
from tensorflow import keras

from tensorflow.keras import Sequential

from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train,y_train),(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
X_test.shape
```

```python
y_train
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(X_train[2])
```

```python
X_train[2]
```

```python
x_train = X_train/255
x_test = X_test/255
```

```python
x_train[2]
```

```python
model = Sequential()

model.add(Flatten(input_shape = (28,28)))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))

#Whenever we are dealing with multiclass classification we use SOFTMAX and number of activation will be equal to Classifications
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy', optimizer = 'Adam', metrics = ['accuracy'])
```

Diff b/w sparse_categorical_crossentropy and sparse_categorical_crossentropy is that we dont have to use one hot encoding 
when we are using sparse

```python
his = model.fit(x_train, y_train, epochs = 10, validation_split = 0.2)
```

```python
y_prob = model.predict(x_test)
```

```python
y_prob
```

Giving probability of each letter, but we want maximum probability value

```python
y_pred = y_prob.argmax(axis = 1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
plt.plot(his.history['loss'])
plt.plot(his.history['val_loss'])
```

```python
plt.plot(his.history['accuracy'])
plt.plot(his.history['val_accuracy'])
```
