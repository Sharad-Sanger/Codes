```python
!pip install apyori
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
retail = pd.read_csv(r"/content/Retail (4).csv", encoding="ISO-8859-1")
```

```python
retail.head(10)
```

```python
retail = retail.drop(['Unnamed: 1', 'Unnamed: 5'], axis=1)
```

```python
retail.shape
```

```python
records = []
for i in range(0, retail.shape[0]):
    records.append([str(retail.values[i, j]) for j in range(0, retail.shape[1])])
```

```python
records[1]
```

```python
association_rules = apriori(records, min_support=0.0055, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
for item in association_results:
    # Extract the rule and its statistics
    rule = item.ordered_statistics[0]
    items_base = ', '.join(list(rule.items_base))
    items_add = ', '.join(list(rule.items_add))
    support = item.support
    confidence = rule.confidence
    lift = rule.lift

    # Print rule details
    print("Rule: {} => {}".format(items_base, items_add))
    print("Support: {:.4f}".format(support))
    print("Confidence: {:.4f}".format(confidence))
    print("Lift: {:.4f}".format(lift))
    print("--------------------------------------")

    # Visualize rule metrics
    plt.figure(figsize=(6, 4))
    labels = [items_base, items_add]
    values = [support, confidence]
    plt.bar(labels, values)
    plt.xlabel("Items")
    plt.ylabel("Metrics")
    plt.title("Association Rule")
    plt.show()
```
