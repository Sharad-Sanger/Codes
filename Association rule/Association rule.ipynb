```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'KB', 'Mushroom', 'Yogurt'],
           ['Pulse', 'Onion', 'Nutmeg', 'KB', 'Mushroom', 'Yogurt'],
           ['Milk', 'Apple', 'KB', 'Mushroom'],
           ['Milk', 'Banana', 'Corn', 'KB', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'KB', 'Mushroom', 'Ice-Cream']]
```

```python
tr = te.fit(dataset).transform(dataset)
```

```python
df = pd.DataFrame(tr, columns = te.columns_)
```

```python
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
ar = apriori(df, min_support = 0.6, use_colnames = True)
```

```python
ar
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
asr = association_rules(ar, metric = 'confidence', min_threshold=0.7)
```
